**Conversion between Star-Delta configurations and generate three phase voltage waveforms for Star and Delta configurations using Python commands**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def star_to_delta(R_A, R_B, R_C):
    R_prod = (R_A * R_B) + (R_B * R_C) + (R_A * R_C)
    R1 = (R_prod / R_C)
    R2 = (R_prod / R_B)
    R3 = (R_prod / R_A)

    return R1, R2, R3

def delta_to_star(R1, R2, R3):
    R_sum = R1 + R2 + R3
    R_A = (R1 * R2) / R_sum
    R_B = (R1 * R3) / R_sum
    R_C = (R2 * R3) / R_sum

    return R_A, R_B, R_C
```

```python
print(star_to_delta(150, 180, 60))
print(delta_to_star(20, 30, 80))
```

```python
# Time array
t = np.linspace(0, 2*np.pi, 1000)

# Star (Y) Configuration Phase Voltages
Vp_mag = 1  # Magnitude of the phase voltages
Va = Vp_mag * np.sin(t)
Vb = Vp_mag * np.sin(t - 2*np.pi/3)
Vc = Vp_mag * np.sin(t + 2*np.pi/3)

# Delta (Δ) Configuration Line Voltages
Vab = Va - Vb
Vbc = Vb - Vc
Vca = Vc - Va

# Plotting
fig, axs = plt.subplots(2, 1, figsize=(10, 8))

# Star (Y) Configuration Plot
axs[0].plot(t, Va, label='Va')
axs[0].plot(t, Vb, label='Vb')
axs[0].plot(t, Vc, label='Vc')
axs[0].set_title('Star (Y) Configuration Phase Voltages')
axs[0].set_xlabel('Time (rad)')
axs[0].set_ylabel('Voltage (V)')
axs[0].legend()
axs[0].grid(True)

# Delta (Δ) Configuration Plot
axs[1].plot(t, Vab, label='Vab')
axs[1].plot(t, Vbc, label='Vbc')
axs[1].plot(t, Vca, label='Vca')
axs[1].set_title('Delta (Δ) Configuration Line Voltages')
axs[1].set_xlabel('Time (rad)')
axs[1].set_ylabel('Voltage (V)')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
